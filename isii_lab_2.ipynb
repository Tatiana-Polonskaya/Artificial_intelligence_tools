### Список наименований средств, соответствующих двум любым документам с требованиями по сертификации

```python
def countDocumnets(string):
    firstDel = string.split(':')
    return len(firstDel[1].split(',')) if len(firstDel) > 1 else 0
```

```python
piece = df['Наименования документов, требованиям которых соответствует средство'].apply(lambda x: str(x).split(':'))
```

```python
df['count_fit_docs'] = df['Наименования документов, требованиям которых соответствует средство'].apply(lambda x: countDocumnets((x)))
```

```python
df_srok_is_bad['Информация об окончании срока технической поддержки, полученная от заявителя'].isnull().any()
```

```python
index_drop = df_srok_is_bad[df_srok_is_bad['Информация об окончании срока технической поддержки, полученная от заявителя'].isnull()].index
```

```python
support_doc[support_doc['Информация об окончании срока технической поддержки, полученная от заявителя']>today]['Наименование\r\nсредства (шифр)']
```
